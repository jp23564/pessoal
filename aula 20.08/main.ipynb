```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('mercado.csv')

cadastro = pd.DataFrame(data)
cadastro
```

```python
# Gráfico de barras
plt.Figure(figsize=(10, 6))

cadastro['Categoria'].value_counts().plot(kind='bar', color=['#413fc0', '#a040b5', '#677a51'])

plt.xlabel('Categorias dos Produtos')

plt.ylabel('Quantidade de Produtos')

plt.title('Quantidade de Produtos por Categoria')

plt.grid(False)

plt.savefig('grafico_bar', bbox_inches='tight')

plt.show()
```

```python
# Gráfico de Pizza
soma_precos_categoria = cadastro.groupby('Categoria')['Preço (R$)'].sum()

plt.figure(figsize=(250, 20))

soma_precos_categoria.plot(kind='pie', autopct='%1.0f%%', color=['#413fc0', '#a040b5', '#677a51'], fontsize=10)

plt.ylabel('')
plt.title('Distribuição de preços por categoria')
plt.savefig('grafico_pizza', bbox_inches='tight')
plt.show()
```

```python
# Gráfico de dispersão
plt.figure(figsize=(15, 6))

plt.grid(True)
sns.scatterplot(data=cadastro, x='Categoria', y='Preço (R$)', hue='Categoria', palette='coolwarm', legend=False, s=100)
plt.title('Relação entre preço e categoria')
plt.savefig('grafico_dispersão', bbox_inches='tight')

plt.show()
```

```python
# Gráfico de caixa
plt.figure(figsize=(20, 10))

plt.grid(True)
sns.boxplot(data=cadastro, x='Categoria', y='Preço (R$)', hue='Categoria', palette='coolwarm', legend=False)
plt.title('Distribuição de preços por categoria')
plt.xlabel('Categoria')
plt.ylabel('Preço')
plt.savefig('grafico_box', bbox_inches='tight')

plt.show()
```
